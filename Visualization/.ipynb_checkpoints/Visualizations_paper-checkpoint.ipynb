```julia
using StatsPlots, DataFrames, CSV, Dates, BSON;
using Statistics:mean, std
```

```julia
# initial dataframes
RL_eval=CSV.read("../out/tracker/1311845_eval_results_v12_24_3001_300_600"*
                        "_all_DDPG_fix_base-256_gn.1_Env-U8-no-layer-norm_1231_3001.csv",DataFrame);
RL_test=CSV.read("../out/tracker/1311845_test_results_v12_24_3001_300_600"*
                        "_all_DDPG_fix_base-256_gn.1_Env-U8-no-layer-norm_1231_3001.csv",DataFrame);
```

```julia
# add run column
RL_eval[!,"run"].= 1;
RL_test[!,"run"].= 1;
```

```julia
for i in 2:10
    RL_eval_new = CSV.read("../out/tracker/1311845_eval_results_v12_24_3001_300_600"*
                        "_all_DDPG_fix_base-256_gn.1_Env-U8-no-layer-norm_123$(i)_3001.csv",DataFrame);
    RL_eval_new[!,"run"].= i;
    RL_eval = vcat(RL_eval, RL_eval_new)
    
    RL_test_new = CSV.read("../out/tracker/1311845_test_results_v12_24_3001_300_600"*
                        "_all_DDPG_fix_base-256_gn.1_Env-U8-no-layer-norm_123$(i)_3001.csv",DataFrame);
    RL_test_new[!,"run"].= i;
    RL_test = vcat(RL_test, RL_test_new)
end
```

```julia
@show names(RL_eval);
```

```julia
# substract the summer violations because of warm weather from comfort violations
RL_eval[!, "T_FH_plus_new"] .= RL_eval[!, "T_FH_plus"] .* (RL_eval[!, "HP"] .> 0);
RL_eval[!, "comfort_new"] .= RL_eval[!, "T_FH_plus_new"] + RL_eval[!, "T_FH_minus"] 
                                + RL_eval[!, "V_HW_plus"] + RL_eval[!, "V_HW_minus"];

RL_test[!, "T_FH_plus_new"] .= RL_test[!, "T_FH_plus"] .* (RL_test[!, "HP"] .> 0);
RL_test[!, "comfort_new"] .= RL_test[!, "T_FH_plus_new"] + RL_test[!, "T_FH_minus"] 
                                + RL_test[!, "V_HW_plus"] + RL_test[!, "V_HW_minus"];
```

```julia
eval_gdf = groupby(RL_eval, [:run]);
test_gdf = groupby(RL_test, [:run]);
```

```julia
RL_eval_kpi = combine(eval_gdf, :rewards => sum => :rewards, :comfort_new => sum => :comfort);
RL_test_kpi = combine(test_gdf, :rewards => sum => :rewards, :comfort_new => sum => :comfort);
```

```julia
sum(RL_eval_kpi.comfort)
```

```julia
RL_eval_kpi_norm = RL_eval_kpi ./ [1, maximum(RL_eval_kpi[!,:rewards]), maximum(RL_eval_kpi[!,:comfort])]';
RL_test_kpi_norm = RL_test_kpi ./ [1, maximum(RL_test_kpi[!,:rewards]), maximum(RL_test_kpi[!,:comfort])]';
```

```julia
scatter(RL_eval_kpi_norm[!,:comfort]', RL_eval_kpi_norm[!,:rewards]', legend=false, markersize=7)
scatter!((RL_eval_kpi_norm[!,:comfort][10],RL_eval_kpi_norm[!,:rewards][10]),  markersize=10, markercolor=:red)

scatter!(RL_test_kpi_norm[!,:comfort]', RL_test_kpi_norm[!,:rewards]', legend=false, markersize=7, marker=:d)
scatter!((RL_test_kpi_norm[!,:comfort][10], RL_test_kpi_norm[!,:rewards][10]),  markersize=10, 
            markercolor=:red, marker=:d)

#plot!( [[RL_eval_kpi_norm[!,:comfort]',  RL_eval_kpi_norm[!,:rewards]'],
#    [RL_test_kpi_norm[!,:comfort]',  RL_test_kpi_norm[!,:rewards]']])

#plot(p1, p2, layout=(1,2), size=(450, 220))
```

```julia
function loadBSON(;idx=NUM_EP, scores_only=false, path="", rng=seed_run)
	if scores_only==true
		BSON.@load "../out/bson/$(path)/DDPG_Shems_v12_$(EP_LENGTH)_$(NUM_EP)_$(L1)_$(L2)_$(case)_$(rng)_scores_$(idx).bson" total_reward score_mean best_run noise_mean
		return total_reward , score_mean, best_run , noise_mean 
	else
		BSON.@load "../out/bson/$(path)/DDPG_Shems_v12_$(EP_LENGTH)_$(NUM_EP)_$(L1)_$(L2)_$(case)_$(rng)_actor_$(idx).bson" actor
		BSON.@load "../out/bson/$(path)/DDPG_Shems_v12_$(EP_LENGTH)_$(NUM_EP)_$(L1)_$(L2)_$(case)_$(rng)_scores_$(idx).bson" total_reward score_mean best_run noise_mean
		return actor , total_reward , score_mean , best_run , noise_mean 
	end
end
```

```julia
score_mean_all = zeros(Float32, (ceil(Int32, NUM_EP/test_every), num_seeds)) 
for i in 1:num_seeds
    test_rng_run = parse(Int, string(seed_ini)*string(i))
    score_mean_all[:,i] = loadBSON(scores_only=true, rng=test_rng_run)[2]
end
# plot_all_scores(ymin = -50, score_mean=score_mean_all) #ymin = -10, 
```
